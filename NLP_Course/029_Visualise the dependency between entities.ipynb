```python
import spacy
from spacy import displacy
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
frasesin = nlp("Tesla's shares have been in free fall since Musk expressed interest in buying Twitter and that was in April.")
```

```python
dep = []

for x in range(1, len(frasesin)):
  dep.append([frasesin[x].text, frasesin[x].dep_, frasesin[x].head.text])
```

```python
dep
```

```python
displacy.render(frasesin, style= "dep", jupyter= True, options= {"distance": 90})
```
